## Splitting Large Geojson(whole state Village Boundary) File into small files(Village) and converting them into Shape file.

#### Required Libraries
* Geopandas
* folium
* pathlib

```python
import geopandas as gdp
import folium
from pathlib import Path
```

###### Import the Shape file here


```python
gdf = gdp.read_file(r"G:\SPATIAL DATA\indian_village_boundaries-master\mh\mh1.geojson")
```

```python
print(gdf.columns)
```

##### TO see the format of the values that stored in Attributes

```python
for name in sorted(gdf["SUB_DIST"].dropna().unique()):
    print(name)
```

#### Filter Data as per the Attributes

#### It filters the data that matches with the village name

```python
matches = gdf[gdf["NAME"] == "Chicholi"]

print(matches[["NAME", "SUB_DIST","DISTRICT", "STATE"]])
```

```python
feature = gdf[
    (gdf["NAME"] == "Chicholi") &  (gdf.index == 8307)
]
```

#### To preview the Selected Attribute

```python
centroid = feature.geometry.iloc[0].centroid

m = folium.Map(
    location=[centroid.y, centroid.x],
    zoom_start=13
)

# Add polygon to map
folium.GeoJson(feature).add_to(m)

m
```

#### To Export as the Shape file

```python
output_folder =r"C:\\Users\\LENOVO\\Desktop\\op"

feature.to_file(
    f"{output_folder}\\Chicholi.shp",
    driver="ESRI Shapefile"
)
```
